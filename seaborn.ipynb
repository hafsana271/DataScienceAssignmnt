```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.violinplot(iris)
plt.show()
```

```python
sns.swarmplot(iris)
plt.show()
```

```python
sns.stripplot(iris)
plt.show()
```

```python
sns.boxplot(iris)
plt.show()
```

```python
sns.barplot(iris)
plt.show()
```

```python
sns.countplot(iris)
plt.show()
```

```python
corr_matrix = iris.drop(columns='species').corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(iris)
plt.show()
```

```python
sns.lineplot(iris)
plt.show()
```
